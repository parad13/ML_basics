```python
#scale
from sklearn import preprocessing
import numpy as np
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
scale = preprocessing.scale(X_train)
scale
```

```python
# StandardScaler
stdscaler= preprocessing.StandardScaler().fit_transform(X_train)
stdscaler
```

```python
#MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
minmax = MinMaxScaler()
minmax.fit(data)
```

```python
minmax.transform(data)
```

```python
minmax.min_ # min - X.min(axis=0) * self.scale_
```

```python
minmax.scale_ # (max - min) / (X.max(axis=0) - X.min(axis=0))
```

```python
minmax.data_min_,minmax.data_max_
```

```python
minmax.data_range_  # same as (data_max_ - data_min_)
```

```python
minmax.n_samples_seen_
```

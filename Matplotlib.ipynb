```python
x=np.arange(0,5*np.pi,0.1)
y=np.sin(x)
plt.title('SIne waveform')
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#sine
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('Sine')

#cosine
plt.subplot(2,1,2)
plt.plot(x,y_cos,'y--')
plt.title('Cosine')

#figure
plt.show()
```

```python
#barplot

x=[2,8,11]
y=[11,16,19]

x1=[3,6,11]
y1=[33,43,19]

plt.bar(x1,y1,color='g')
plt.bar(x,y,color='b')
plt.title('BAr')
plt.xlabel("xaxis")
plt.ylabel('yaxis')

plt.show()
```

```python
#histogram
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a)
plt.title('histogram')
plt.xlabel("xaxis")
plt.ylabel('yaxis')

plt.show()
```

```python
#boxplot
data=[np.random.normal(0,std,8) for std in range(1,4)]
plt.boxplot(data,vert=False,patch_artist=False)
plt.xlabel('x')
plt.ylabel('y')
plt.title("boxplot")
plt.show()
plt.savefig('box1.png')
```

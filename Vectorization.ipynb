```python
import numpy as np
a = np.array([1, 2, 3, 4])
print(a)
```

```python
import time

#Creation of two random array using np
a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a, b)
toc = time.time()

print(c)
print("Vectorized version: " + str(1000*(toc-tic)) + "ms")

c = 0
tic = time.time()
for _ in range(1000000) :
  c += a[_]*b[_]
toc = time.time()

print(c)
print("For loop version: " + str(1000*(toc-tic)) + "ms")

# c = 0
# tic = time.time()
# [[c += a[_]*b[_]] for _ in range(1000000)]
# # for _ in range(1000000) :
# #   c += a[_]*b[_]
# toc = time.time()

# print(c)
# print("List comprehension version: " + str(1000*(toc-tic)) + "ms")
```
